```python
ip=iris.drop('flower_type',axis=1)
ip.head()
```

```python
#k-mean clustering
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km.fit(ip)
```

```python
km_pred = km.predict(ip)
km_pred
```

```python
iris['prediction']= km_pred
iris
```

```python
centroid=km.cluster_centers_
centroid
```
